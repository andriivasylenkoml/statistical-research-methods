```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import cumfreq

x2 = np.array([2.6, 2.5, 0.8, 2.8, 2.4, 2.1, 2.3, 3.4, 1.7, 1.7, 2.2, 1.3, 3.4, 2.3, 1.6, 1.4])

n_classes = 5
hist, bin_edges = np.histogram(x2, bins=n_classes)
class_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.hist(x2, bins=n_classes, edgecolor="black", alpha=0.7)
plt.title("Histogram")

plt.subplot(1, 2, 2)
plt.plot(class_centers, hist, marker="o", linestyle="-", color="blue")
plt.title("Polygon of frequencies")
plt.grid(True)

cumfreqs = cumfreq(x2, numbins=n_classes)

plt.figure(figsize=(10, 6))
x_cdf = cumfreqs.lowerlimit + np.linspace(0, cumfreqs.binsize * cumfreqs.cumcount.size, cumfreqs.cumcount.size)
plt.plot(x_cdf, cumfreqs.cumcount / len(x2), marker="o", linestyle="-", color="green")
plt.title("Integral distribution function")
plt.grid(True)

plt.show()
```
